# Run locally

## Set-up

````python
import requests
import json

import pandas as pd
````

## 1. Gunicorn

Start app by running the following commands in terminal:
    
```bash
export $(xargs <.env)

gunicorn app.fpl_team_selector:app --bind 0.0.0.0:5000
```

````python
payload = {
    "previous_gw": "28",
    "season": "2019-20",
    "previous_team_selection_path": "s3://fpl-analysis-data/gw_team_selections/season=2019-20/gw=28/gw28_v3_lstm_team_selections.parquet", 
    "budget": "100.4", 
    "available_chips": [],  #["wildcard"],
    "available_transfers": "2"
}
````

````python
response = requests.get("http://0.0.0.0:5000/retro", json=payload)
````

````python
response_as_dict = json.loads(response.content)
````

````python
response_as_dict.keys()
````

````python
response_as_dict['transfers']
````

````python
pd.DataFrame(response_as_dict['team_selected'])
````
